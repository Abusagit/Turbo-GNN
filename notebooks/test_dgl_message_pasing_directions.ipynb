```python
import dgl

import torch



```

```python
src = torch.tensor([0, 1, 0, 1], dtype=torch.int64)
dst = torch.tensor([1, 0, 2, 2], dtype=torch.int64)

features = torch.tensor([
    [1, 1],
    [2, 1],
    [1, 2]
], dtype=torch.float32)


graph = dgl.graph((src, dst))
graph
```

```python
in_degrees = graph.in_degrees()
out_degrees = graph.out_degrees()


in_degrees, out_degrees
```

```python
graph.adj().to_dense()
```

```python
graph.adj().to_dense() / in_degrees.unsqueeze(1)
```

```python
torch.diag(1 / in_degrees) @ graph.adj().to_dense().T @ features
```

```python
dgl.ops.copy_u_mean(graph, features)
```

```python
conv = dgl.nn.pytorch.GraphConv(weight=False, bias=False, in_feats=2, out_feats=2, norm='both')
```

```python
conv(graph, features)
```
